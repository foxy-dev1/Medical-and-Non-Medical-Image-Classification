```python
import os
import shutil

def safe_merge(src, dst):
    os.makedirs(dst, exist_ok=True)
    for root, _, files in os.walk(src):
        if len(files)>0:
            dir_name = os.path.basename(root)
            for file in files:
                src_file = os.path.join(root, file)
                dst_file = os.path.join(dst, file)
    
                if os.path.exists(dst_file):
                    base, ext = os.path.splitext(file)
                    i = 1
                    while os.path.exists(dst_file):
                        dst_file = os.path.join(dst, f"{dir_name}_{base}_{i}{ext}")
                        i += 1
    
                shutil.copy2(src_file, dst_file)
```

### make datasets

```python
from PIL import Image
import os

def verify_images(folder_path):
    for root, _, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            try:
                img = Image.open(file_path)
                img.verify()  
            except Exception as e:
                print(f"Deleting bad image: {file_path}")
                os.remove(file_path)

verify_images("/kaggle/working/test")
```

```python
train_images_path = "/kaggle/working/train"
test_images_path = "/kaggle/working/test"
device = 'cuda' if torch.cuda.is_available() else 'cpu'


train_transform = transforms.Compose([transforms.Resize(256),transforms.CenterCrop(224),
                                transforms.RandomRotation(90),
                                transforms.RandomHorizontalFlip(),
                                transforms.Grayscale(num_output_channels=3),
                                transforms.ToTensor(),
                                transforms.Normalize([0.485,0.456,0.406],
                                                    [0.229,0.224,0.225])]        
                        )

test_transform = transforms.Compose([transforms.Resize(256),
                                     transforms.CenterCrop(224),
                                transforms.Grayscale(num_output_channels=3),
                                transforms.ToTensor(),
                                transforms.Normalize([0.485,0.456,0.406],
                                                    [0.229,0.224,0.225])]        
                        )

train_dataset = datasets.ImageFolder(train_images_path,transform=train_transform)
test_dataset = datasets.ImageFolder(test_images_path,transform=test_transform)

# train_all_data = []
# for img,label in train_dataset:
#     train_all_data.append((img.to(device),torch.tensor(label,device=device)))

# test_all_data = []
# for img,label in test_dataset:
#     test_all_data.append((img.to(device),torch.tensor(label,device=device)))
    

train_dataloader = torch.utils.data.DataLoader(train_dataset,batch_size=64,shuffle=True)
test_dataloader = torch.utils.data.DataLoader(test_dataset,batch_size=64,shuffle=True)
```

```python
import torchvision.models as models
import torch.nn as nn
import torch

vgg_model = models.vgg16(weights=models.VGG16_Weights.DEFAULT)

for param in vgg_model.features.parameters():
    param.requires_grad = False


in_features = vgg_model.classifier[-1].in_features

vgg_model.classifier[-1] = nn.Linear(in_features,5)
```

```python
# vgg_model
```

```python
for input,label in train_dataloader:
    print(input.shape,label.shape)
    break
```

```python
loss_fn = nn.CrossEntropyLoss()
optim = torch.optim.Adam(vgg_model.parameters(),lr=0.0001)

if torch.cuda.device_count()>1:
    print("using multi gpu's")
    vgg_model = nn.DataParallel(vgg_model)
    
vgg_model.to(device)
```

```python
import time

start_time = time.perf_counter()

epochs = 140
best_model_state = None
best_model_loss = float('inf')

for epoch in range(epochs):
    vgg_model.train()
    running_loss = 0

    for data,labels in train_dataloader:
        data,labels = data.to(device),labels.to(device)
        
        optim.zero_grad()
        outputs = vgg_model(data)
        loss = loss_fn(outputs,labels)
        loss.backward()
        running_loss += loss.item()
        optim.step()
    
    if running_loss < best_model_loss:
        best_model_loss = running_loss
        if isinstance(vgg_model,nn.DataParallel):
            best_model_state = vgg_model.module.state_dict()
        else:
            best_model_state = vgg_model.state_dict()

    
    # if epoch % 10 ==0 or epoch==epochs-1:
    print(f"{epoch} -> loss: {running_loss:.3f}, best loss -> {best_model_loss:.3f}")
    
torch.save(best_model_state,"best_model.pth")        
end_time = time.perf_counter()

elapsed_time = end_time - start_time
print(f"training time -> {elapsed_time / 60:.2f} minutes")
```

```python
vgg_model.eval()
correct = 0
total = 0
with torch.no_grad():
    for test_data,test_labels in test_dataloader:
        test_data,test_labels = test_data.to(device),test_labels.to(device)

        output = vgg_model(test_data)        
        _, predicted = torch.max(output, 1)
        total += len(predicted)
        correct += (predicted==test_labels).sum().item()

print(f"acc -> {(correct/total):.2f}")
```
